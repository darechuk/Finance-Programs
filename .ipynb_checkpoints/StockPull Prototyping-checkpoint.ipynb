# Using Alpha Vantage API to pull Financial Data for Analysis

```python
import pandas as pd
import datetime
from datetime import datetime as dt, timedelta
from alpha_vantage.timeseries import TimeSeries
from alpha_vantage.cryptocurrencies import CryptoCurrencies
import os
import pickle
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter, MinuteLocator, YearLocator, MonthLocator
from matplotlib.finance import candlestick_ohlc

%matplotlib inline
plt.style.use('seaborn-paper')
```

### Import Alpha Vantage API Key for text file

```python
def key_import(filename):
    with open(filename, 'r') as txtfile:
        APIKey = txtfile.read()
    return(APIKey)

#Get API from AlphaVantageAPIKey.txt in Notebook Folder
APIKey = key_import('AlphaVantageAPIKey.txt')
```

### Function for Intraday pull

```python
def pull_Intraday(ticker):        
    TS = TimeSeries(key=APIKey, output_format='pandas')
    result = TS.get_intraday(symbol=ticker,interval='1min',outputsize='full')
    return(result)
```

### Function for Daily Adjusted pull

```python
def pull_DailyAdj(ticker):        
    TS = TimeSeries(key=APIKey, output_format='pandas')
    result = TS.get_daily_adjusted(symbol=ticker, outputsize='full')
    return(result)
```

### Function for Intraday Digital Currency pull

```python
def pull_IntradayDC(ticker):        
    CC = CryptoCurrencies(key=APIKey, output_format='pandas')
    result = CC.get_digital_currency_intraday(symbol=ticker, market='USD')
    return(result)
```

### Function for Daily Digital Currency pull

```python
def pull_DailyDC(ticker):        
    CC = CryptoCurrencies(key=APIKey, output_format='pandas')
    result = CC.get_digital_currency_daily(symbol=ticker, market='USD')
    return(result)
```

### Functions to run pulls with time of day check for backing up data

```python
def run_intraday(symbol):
    pname = symbol + str(dt.now().date()) + ".p"
    endtime = dt.combine(dt.now().date(), datetime.time(16,0,0))
    starttime = dt.combine(dt.now().date(), datetime.time(8,0,0))
    
    if starttime < dt.now() < endtime:
        df = pull_Intraday(symbol)
    else:
        if os.path.isfile(pname):
            with open(pname, "rb") as pfile:
                df = pickle.load(pfile)
        else:
            df = pull_Intraday(symbol)
            pickle.dump(df,open(pname, "wb"))
    return(df)

def run_dailyadj(symbol):
    pname = symbol + str(dt.now().date()) + "Daily" + ".p"
    endtime = dt.combine(dt.now().date(), datetime.time(16,0,0))
    starttime = dt.combine(dt.now().date(), datetime.time(8,0,0))
    if os.path.isfile(pname):
        with open(pname, "rb") as pfile:
            df = pickle.load(pfile)
    else:
        df = pull_DailyAdj(symbol)
        pickle.dump(df,open(pname, "wb"))
    return(df)
```

### Run Stuff

```python
symbol = "AAPL"
data, meta = run_intraday(symbol)

datelist = []
for i in range(0,len(data.index)): #Convert Date index in dataframe from string to datetime.datetime object
    data.index.values[i] = dt.strptime(data.index.values[i],'%Y-%m-%d %H:%M:%S')
    if not data.index.values[i].date() in datelist: #Populate datelist with dates in Index
        datelist.append(data.index.values[i].date())

#Create subdataset for plotting
i = len(datelist) - 1
starttime = dt.combine(datelist[i], datetime.time(8,0,0))
endtime = dt.combine(datelist[i], datetime.time(16,0,0))
subdata = data.loc[starttime:endtime]
x = mdates.date2num(subdata.index.values) #Create plot x-axis by converting list of datetime.datetime index values to float list

#Plot
fig1 = plt.figure(1, dpi=150)
fig1.suptitle(symbol + " (" + str(datelist[i]) + ")", fontsize=12)
plt.subplots_adjust(hspace=0)

ax1 = plt.subplot2grid((8,1),(0,0), rowspan=5, colspan=1)
plt.plot(x,subdata['4. close'])
ax1.axhline(y = subdata['4. close'][len(x)-1], ls = ':', color='r')
plt.xlim(x.min(),x.max())
plt.ylabel("Price (USD)")
#plt.gca().xaxis.set_major_locator(MinuteLocator(byminute=range(0,60,30)))
plt.gca().xaxis.set_major_formatter(DateFormatter('%H:%M'))

ax2 = plt.subplot2grid((8,1),(6,0), rowspan=2, colspan=1)
plt.bar(x,subdata['5. volume'],0.0025)
plt.xlabel("Timestamp")
plt.ylabel("Volume")
plt.xlim(x.min(),x.max())
#plt.gca().xaxis.set_major_locator(MinuteLocator(byminute=range(0,60,30)))
plt.gca().xaxis.set_major_formatter(DateFormatter('%H:%M'))
```

### Run More Stuff

```python
symbol = "AAPL"
startdate = dt(2017,9,1)
data, meta = run_dailyadj(symbol)

#Create subdataset for plotting
subdata = data.loc[str(startdate):]

for i in range(0,len(subdata.index)): #Convert Date index in dataframe from string to datetime.datetime object to matplotlib float
    subdata.index.values[i] = dt.strptime(subdata.index.values[i],'%Y-%m-%d')
    subdata.index.values[i] = mdates.date2num(subdata.index.values[i])
    
#Create series for Exponential weighted mean
y5 = subdata['5. adjusted close'].ewm(span=5).mean()
y25 = subdata['5. adjusted close'].ewm(span=25).mean()

ycandle = subdata.iloc[:,0:4]
ycandle.reset_index(inplace=True)
ycandle.columns = ['t','open','high','low','close']

#Generate Plot
fig1 = plt.figure(1, dpi=150)
fig1.suptitle(symbol, fontsize=12)
plt.subplots_adjust(hspace=0)

ax1 = plt.subplot2grid((6,1),(0,0), rowspan=6, colspan=1)
plt.scatter(subdata.index,subdata['5. adjusted close'],color='black',s=10, label = "Adjusted Close")
plt.plot(subdata.index,y5,'--', label = "Adjusted Close EWMA N=5")
plt.plot(subdata.index,y25,'-.', label = "Adjusted Close EWMA N=25")
plt.xlim(subdata.index.min(),subdata.index.max())
plt.ylabel("Price (USD)")
plt.xlabel("Date")
cnd = candlestick_ohlc(ax1, ycandle.values, width=0.75, colorup='g', colordown='r', alpha=1.0)
#plt.gca().xaxis.set_major_locator(MonthLocator(range(1, 13), bymonthday=1, interval=1))
plt.gca().xaxis.set_major_formatter(DateFormatter('%b-%y'))
lgd = ax1.legend()


```
